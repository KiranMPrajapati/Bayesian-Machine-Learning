```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12, 10)
```

```python
def posterior_dirichlet_categorical_conditional_distribution(count, alpha, k=3):
    C = np.sum(count)
    A = np.sum(alpha)
    num = count[k] + alpha[k]
    denom = C + A - 1
    
    return num / denom
```

```python
count = [3,4,5,6,7,2,8, 9]
alpha = [0.8, 1, 3, 2.2, 1.6, 1.9, 2, 2]
```

```python
x = np.linspace(0, len(count)-1, len(count))
```

```python
y = []
for i in range(len(count)):
    y.append(posterior_dirichlet_categorical_conditional_distribution(count, alpha, i))
```

```python
plt.plot(x, y, label=f"alpha={alpha}, count={count}")
plt.legend()
```
